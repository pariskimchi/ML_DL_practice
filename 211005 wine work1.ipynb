## Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
# Function to calculate Cross validation 
def exec_kfold(clf, folds=5):
    kfold = KFold(n_splits=folds)
    scores=[]
    for iter_count, (train_index, test_index) in enumerate(kfold.split(x_df)):
        x_train, x_test = x_df.values[train_index], x_df.values[test_index]
        y_train, y_test = y_df.values[train_index], y_df.values[test_index]
        clf.fit(x_train, y_train)
        predictions = clf.predict(x_test)
        accuracy = accuracy_score(y_test, predictions)
        scores.append(accuracy)
        print('êµì°¨ê²€ì¦:{0} ì •í™•ë„ :{1:.4f}'.format(iter_count, accuracy))
        
    mean_score = np.mean(scores)
    print('í‰ê·  ì •í™•ë„:{0:.4f}'.format(mean_score))
    
```

### êµì°¨ê²€ì¦ (StratifiedKFold)

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=5)
num=0
```

```python
for train_index, test_index in skf.split(x_df, y_df):
    num+=1
    label_train = y_df.iloc[train_index]
    label_test = y_df.iloc[test_index]
    print('### êµì°¨ê²€ì¦: {0}'.format(num))
    print('í•™ìŠµë ˆì´ë¸” ë°ì´í„° ë¶„í¬: \n', label_train.value_counts())
    print('ê²€ì¦ë ˆì´ë¸” ë°ì´í„° ë¶„í¬:\n', label_test.value_counts())
```

```python
# kfold + shuffle 
def exec_kfold_shuffle(clf, folds=5):
    kfold = KFold(n_splits=folds, shuffle=True)
    scores=[]
    for iter_count, (train_index, test_index) in enumerate(kfold.split(x_df)):
        x_train, x_test = x_df.values[train_index], x_df.values[test_index]
        y_train, y_test = y_df.values[train_index], y_df.values[test_index]
        clf.fit(x_train, y_train)
        predictions = clf.predict(x_test)
        accuracy = accuracy_score(y_test, predictions)
        scores.append(accuracy)
        print('êµì°¨ê²€ì¦:{0} ì •í™•ë„:{1:.4f}'.format(iter_count, accuracy))
        
    mean_score = np.mean(scores)
    print('í‰ê·  ì •í™•ë„:{0:.4f}'.format(mean_score))
        
```

```python
exec_kfold_shuffle(dt_clf, 50)
```
